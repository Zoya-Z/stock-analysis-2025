```python
import yfinance as yf #collect data from yfinance
import pandas as pd #data anlaysis
import matplotlib.pyplot as plt #data visualisation
import seaborn as sns #data visualisation
```

```python
start_date ='2025-01-01'
end_date ='2025-10-01'
```

```python
circle = yf.download('CRCL', start_date, end_date)
```

```python
circle.head() #top.5 rows
```

```python
circle.tail() #bottmo.5 rows
```

```python
circle.columns
```

```python
circle ['Close'].plot(figsize = (16,8))
plt.xlabel('date')
plt.ylabel('$price')
plt.title('circle stock price')
plt.savefig('pic.png')
```

```python
tickers = ['META','AMZN','AAPL','NFLX','GOOG']
portfolio = yf.download(tickers, start_date,end_date)
```

```python
close = portfolio['Close']
close.head() #
```

```python
portfolio['Close'].plot(figsize = (16,8))
```

```python
portfolio['Close'].pct_change().corr()
```

```python
corrMatrix = portfolio['Close'].pct_change().corr()
sns.heatmap(corrMatrix,annot = True)
plt.savefig('gip.png')
```
